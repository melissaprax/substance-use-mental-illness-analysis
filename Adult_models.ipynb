```python
# Import dependencies
import pandas as pd
import numpy as np
from pycaret.classification import *
```

## Section I: Young adults

```python
file_path = 'data/subset_three_targets.csv'

# Load the file 
df = pd.read_csv(file_path)
```

```python
# Create Young ADULT dataframe (CATAG3 == 2)
YOUNGADULT = df[df['CATAG3'].isin([2])]

# Drop the YODSMMDE and DEP columns from young ADULT dataframe
YOUNGADULT = YOUNGADULT.drop(['YODSMMDE', 'DEP'], axis=1)

# Create ADULT dataframe (CATAG3 == 3)
ADULT = df[df['CATAG3'].isin([3, 4])]

# Drop the YODSMMDE and DEP columns from ADULT dataframe
ADULT = ADULT.drop(['YODSMMDE', 'DEP'], axis=1)
```

```python
# import pycaret classification and init setup
s = setup(YOUNGADULT, target = 'ADSMMDEA', session_id = 123)
```

```python
# import ClassificationExperiment and init the class
from pycaret.classification import ClassificationExperiment
exp = ClassificationExperiment()
```

```python
# check the type of exp
type(exp)
```

```python
# init setup on exp
exp.setup(YOUNGADULT, target = 'ADSMMDEA', session_id = 123)
```

```python
# Compare baseline models
best = compare_models()
```

```python
# Compare baseline models
best = compare_models()
```

```python
# Compare models using OOP
exp.compare_models()
```

```python
# plot confusion matrix
plot_model(best, plot = 'confusion_matrix')
```

```python
# plot AUC
plot_model(best, plot = 'auc')
```

```python
# plot feature importance
plot_model(best, plot = 'feature')
```
